```python
# Handling outliers with IQR
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# Apply outlier removal
numeric_cols_prod = df_productivity.select_dtypes(include=['float64', 'int64']).columns
numeric_cols_work = df_work_life.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols_prod:
    df_productivity = remove_outliers(df_productivity, col)

for col in numeric_cols_work:
    df_work_life = remove_outliers(df_work_life, col)
```

The code applies the **IQR-based outlier removal** method to all numerical columns in the **Productivity** and **Work-Life Balance** datasets. First, it identifies numerical columns in each dataset using the `select_dtypes` function, which selects columns of type `float64` and `int64`. Then, for each numerical column, the previously defined `remove_outliers` function is applied. This function computes the IQR, determines lower and upper bounds for acceptable values, and filters out data points falling outside these bounds. As a result, both datasets are cleansed of outliers, ensuring that the data is within a reasonable range and reducing the potential distortion caused by extreme values in further analysis or modeling.
